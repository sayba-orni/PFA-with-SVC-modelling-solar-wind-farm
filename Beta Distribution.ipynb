```python
def Beta(alpha, beta, s):
    B=(scipy.special.gamma(alpha)*scipy.special.gamma(beta))/scipy.special.gamma(alpha+beta);
    return (1/B)*(s**(alpha-1))*((1-s)**(beta-1))
```
